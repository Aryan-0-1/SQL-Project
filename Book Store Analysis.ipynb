# Database Creation

```python
# Create Database for Online Book Store
conn = sqlite3.connect("Online Book Store.db")
cursor = conn.cursor()
```

```python
# Creating Table for Books Information in Database
cursor.execute("""
    DROP TABLE IF EXISTS Books
""")
conn.commit()

cursor.execute("""
    CREATE TABLE Books(
    Book_ID INTEGER PRIMARY KEY,
    Title VARCHAR(100),
    Author VARCHAR(100),
    Genre VARCHAR(50),
    Published_Year INTEGER,
    Price NUMERIC(10, 2),
    Stock INTEGER
    );
""")
conn.commit()
```

```python
# Adding Books Information in Database
df = pd.read_csv('C:\\Users\\aryan\\OneDrive\\Desktop\\Data Science\\Projects\\SQL\\Books.csv')
df.to_sql(
    "Books",      
    conn,
    if_exists="replace",
    index=False
)
```

```python
# Creating Table for Customers Information in Database
cursor.execute("""
    DROP TABLE IF EXISTS Customers
""")
conn.commit()

cursor.execute("""
    CREATE TABLE Customers(
    Customer_ID INTEGER PRIMARY KEY,
    Name VARCHAR(100),
    Email VARCHAR(100),
    Phone INTEGER,
    City VARCHAR(50),
    Country VARCHAR(50)
    );
""")
conn.commit()
```

```python
# Adding Customers Information in Database
df = pd.read_csv('C:\\Users\\aryan\\OneDrive\\Desktop\\Data Science\\Projects\\SQL\\Customers.csv')
df.to_sql(
    "Customers",
    conn,
    if_exists="replace",
    index=False
)
```

```python
# Creating Table for Orders Information in Database
cursor.execute("""
    DROP TABLE IF EXISTS Orders
""")
conn.commit()

cursor.execute("""
    CREATE TABLE Orders(
    Order_ID INTEGER PRIMARY KEY,
    Book_ID INTEGER,
    Customer_ID INTEGER,
    Order_Date DATE,
    Quantity INTEGER,
    Total_Amount NUMERIC(10, 2)
    );
""")
conn.commit()
```

```python
# Adding Orders Information in Database
df = pd.read_csv('C:\\Users\\aryan\\OneDrive\\Desktop\\Data Science\\Projects\\SQL\\Orders.csv')
df.to_sql(
    "Orders",
    conn,
    if_exists="replace",
    index=False
)
```

# Advance Queries

```python
# T1 :- Retrieve the total number of books sold from each genre
query = """
    SELECT B.Genre, SUM(O.Quantity) AS 'Total Books Sold' FROM Books B
    JOIN Orders O
    ON B.Book_ID = O.Book_ID
    GROUP BY B.Genre
"""

df = pd.read_sql_query(query, conn)
df
```

```python
# T2 :- Find the average price of books in the "Fantasy" genre
query = """
    SELECT Genre,AVG(Price) AS 'Average Price' FROM Books WHERE Genre = "Fantasy"
"""

df= pd.read_sql_query(query, conn)
df
```

```python
# T3 :- List customers who have placed at least 2 orders
query = """
    WITH count_table AS(SELECT Customer_ID ,COUNT(Customer_ID) AS 'Number_of_Orders' FROM Orders GROUP BY Customer_ID)
    
    SELECT T.Customer_ID, C.Name, T.Number_of_Orders FROM Customers C
    RIGHT JOIN
    (SELECT Customer_ID, Number_of_Orders FROM count_table
    WHERE Number_of_Orders >= 2) T
    ON C.Customer_ID = T.Customer_ID
"""


"""
SELECT o.customer_id, c.name, COUNT(o.Order_id) AS ORDER_COUNT
FROM orders o
JOIN customers c ON o.customer_id=c.customer_id
GROUP BY o.customer_id, c.name
HAVING COUNT(Order_id) >=2;
"""

df = pd.read_sql_query(query, conn)
df
```

```python
# T4 :- Find the most frequently ordered book
query = """
    WITH count_table AS(SELECT Book_ID ,COUNT(Order_ID) AS 'Number_of_Orders' FROM Orders GROUP BY Book_ID)
    
    SELECT B.*,T.Number_of_Orders FROM Books B
    RIGHT JOIN
    (SELECT Book_ID,Number_of_Orders FROM count_table
    WHERE Number_of_Orders == (SELECT MAX(Number_of_Orders) FROM count_table)) T
    ON B.Book_ID = T.Book_ID;
"""

df = pd.read_sql_query(query, conn)
df
```

```python
# T5 :- Show the top 3 most expensive books of 'Fantasy' Genre
query = """
    SELECT * FROM Books WHERE Genre == 'Fantasy'
    ORDER BY Price DESC LIMIT 3
"""

df = pd.read_sql_query(query, conn)
df
```

```python
# T6 :- Retrieve the total quantity of books sold by each author
query = """
    SELECT B.Author,SUM(O.Quantity) AS 'Total_Quantity_of_Books_Sold' FROM Books B
    JOIN Orders O
    ON B.Book_ID = O.Book_ID
    GROUP BY B.Author
"""
df = pd.read_sql_query(query, conn)
df
```

```python
# Authors with more than 1 Book
query = """
    WITH count_table AS(SELECT Author, COUNT(Book_ID) AS 'Total_Quantity' FROM Books
    GROUP BY Author)

    SELECT * FROM count_table WHERE Total_Quantity >1
"""

df = pd.read_sql_query(query, conn)
df
```

```python
# T7 :- List the cities where customers who spent over $30 are located
query = """
    WITH spent_table AS (
                         SELECT Customer_ID, SUM(Total_Amount) AS 'Total_Spent' FROM Orders
                         GROUP BY Customer_ID
                         )

    SELECT DISTINCT(City) FROM Customers
    WHERE Customer_ID IN (SELECT Customer_ID FROM spent_table WHERE Total_Spent > 30 )
"""

df = pd.read_sql_query(query, conn)
df
```

```python
# T8 :- Find the customer who spent the most on orders
query = """
    WITH spent_table AS (
                         SELECT Customer_ID, SUM(Total_Amount) AS 'Total_Spent' FROM Orders
                         GROUP BY Customer_ID
                         )
                
    SELECT * FROM Customers C
    RIGHT JOIN (SELECT Customer_ID, MAX(Total_Spent) FROM spent_table) T
    ON C.Customer_ID = T.Customer_ID
"""

df = pd.read_sql_query(query, conn)
df
```

```python
# T9 :- Calculate the stock remaining after fulfilling all orders
query = """
     WITH books_ordered AS (
                         SELECT Book_ID, SUM(Quantity) AS 'Total_Books_Ordered' FROM Orders
                         GROUP BY Book_ID
                         )
                         
    SELECT B.*, COALESCE((B.Stock - O.Total_Books_Ordered),B.Stock) AS 'Remaning_Stock' FROM Books B
    LEFT JOIN (SELECT * FROM books_ordered) O
    ON B.Book_ID = O.Book_ID
"""

df = pd.read_sql_query(query, conn)
df
```
